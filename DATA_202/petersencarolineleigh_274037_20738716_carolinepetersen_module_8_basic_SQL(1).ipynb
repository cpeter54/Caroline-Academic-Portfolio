## Exercise 1
Use the `tracks` table to determine how many albums each artist has made. Your query should return each unique `Composer` and the number of albumns they created. 

- Alias the number of albums `NumAlbums`
- Check to make sure `Composer` is not null
- Sort the results in descending order of number of album%load_ext sql


```python
%%sql
SELECT Composer, COUNT(AlbumID) AS NumAlbums
FROM tracks 
WHERE Composer IS NOT NULL
GROUP BY Composer
ORDER BY NumAlbums DESC
```

## Exercise 2
What day of the week do most people buy music on? Write a query to return the day of the week and the number of invoices created on that day of the week. 

- Alias the day of the week to have the column name `DayOfWeek`
- Alias the number of invoices created on that day to be `NumInvoice`

Strategy:
1. Find the day of the week for each invoice in the `invoices` table
2. Group by day of the week
3. Count the number of items for each group

```python
%%sql
SELECT strftime('%w', InvoiceDate) AS DayOfWeek, COUNT(InvoiceDate) AS NumInvoice
FROM invoices
GROUP BY DayOfWeek
```

## Exercise 3
Use the `tracks` table to create a table showing statistics on the lengths of songs in albums that are produced by composers.

Your query should return:
- Each unique `Composer`
- Each unique `AlbumID` by `Composer`
- The minimum song length of each album aliased to be `MinSongLength`
- The maximum song length of each album aliased to be `MaxSongLength`
- The average song length of each album aliased to be `MeanSongLength`
- Perform a check to ensure `Composer` is not null
- Filter groups to use albulms with more than one song
- Sort the results by average song length in ascending order

You can convert to minutes, but you do not need to.

```python
%%sql
SELECT Composer, AlbumID, MIN(Milliseconds)AS MinSongLength, MAX(Milliseconds)AS MaxSongLength, AVG(Milliseconds) AS MeanSongLength FROM tracks
WHERE Composer IS NOT NULL
GROUP BY Composer, AlbumID
HAVING COUNT(AlbumId)>1
ORDER BY MeanSongLength ASC
```
